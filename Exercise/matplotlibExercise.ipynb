```python
# pytohon 绘图库matplotlib
import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0,2*np.pi,50)
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x), x, np.cos(x))
plt.show()
```

```python
plt.plot(x, np.sin(x),'g-', x, np.cos(x),'r--')
plt.show()
```

```python
plt.subplot(2,1,1)  # (行，列，活跃区)  
plt.plot(x, np.sin(x),'g-')
plt.subplot(2,1,2)  # (行，列，活跃区)  
plt.plot(x, np.cos(x),'r--')
plt.show()
```

```python
# 彩色映射散点图
a = np.random.rand(1000)
b = np.random.rand(1000)
size = np.random.rand(1000) * 50
color = np.random.rand(1000)
# 调用 scatter() 函数并传入两个分别代表 x 坐标和 y 坐标的数组a 和b 
# size 点的大小，color点的颜色
plt.scatter(a,b,size,color) 
plt.colorbar()
plt.show()
```

```python
#直方图
c = np.random.rand(1000)
plt.hist(c,50)
plt.show()
```

```python
# 设置图表的标题，标签和图例
x1 = np.linspace(0, 2*np.pi,50)
plt.plot(x1,np.sin(x1),'r-.', label='Sin(x)')
plt.plot(x1,np.cos(x1),'g-.', label='Cos(x)')
plt.legend()  #展示图例
plt.xlabel('Rads')
plt.ylabel('Amplitude')
plt.title('Sin and Cos Waves')
plt.show()
```
